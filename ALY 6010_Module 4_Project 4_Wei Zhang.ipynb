## Part 2      
Using the random sample from Part 1, complete the following analysis of the data.      
1.Calculate the sample mean, sample standard deviation, sample size and sampling (standard) error.      

```python
df_s1=pd.DataFrame({'Value':[arr1.mean(),np.var(arr1),len(arr1),np.std(arr1)]},
                  index=['Sample Mean','Sample Variance','Sample Size',
                         'Sample Standard deviation'])
df_s1
```

2.Use the sample to create 92%, 96% and 98% confidence intervals for the population mean.

```python
#use Z table to find Z
def confidence(arr,z):
    n=len(arr)
    m=mean(arr)
    std=np.std(df_usp['International Migration Rate'])
    
    E=(z*std)/(n**0.5)
    start = m - E
    end = m + E
    E=round(E,6)
    return [start,end,-E,E]

con1=confidence(arr1,1.75)
con2=confidence(arr1,2.05)
con3=confidence(arr1,2.33)


df_s2=pd.DataFrame({'Confidence Level CI':[0.92,0.96,0.98],
                   'z or t value corresponding to the CI level':[1.75,2.05,2.33],
                   'Margin of Error':[con1[2:],con2[2:],con3[2:]],
                   'CI Lower Limit':[con1[0],con2[0],con3[0]],
                   'CI Upper Limit':[con1[1],con2[1],con3[1]],
                   'CI Width':[con1[1]-con1[0],con2[1]-con2[0],con3[1]-con3[0]]},
                  index=[0,1,2])
df_s2
```

3.For each of the confidence intervals in step 2, use the information already calculated to determine the sample size needed to estimate the population mean to within .05 units.     

```python
def sampleSize(z,std,E):
    n=((z*std)/E)**2
    return n

n1=int(sampleSize(1.75,np.std(df_usp['International Migration Rate']),0.05))+1
n2=int(sampleSize(2.05,np.std(df_usp['International Migration Rate']),0.05))+1
n3=int(sampleSize(2.33,np.std(df_usp['International Migration Rate']),0.05))+1

df_s3=pd.DataFrame({'Confidence Level CI':[0.92,0.96,0.98],
                  'Desired Margin of Error':[[-0.05,0.05],[-0.05,0.05],[-0.05,0.05]],
                  'Minimum Sample Size Needed':[n1,n2,n3]},
                  index=[0,1,2])
df_s3
```
