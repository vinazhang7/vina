## Task 1:

**Create a Pareto chart of the top 10 states with the highest average number of vehicles per household in 2016.**    
    
    
    

```python
df=uscars[['Jusidiction','2016 Vehicles per Household']]
df.head(5)
```

```python
df['states']=df['Jusidiction'].apply(lambda s:s.split(',')[-1])
df.head(10)
```

```python
df_1=df.groupby('states').mean().sort_values(by='2016 Vehicles per Household',ascending=False).head(10)
df_1['percentage']=df_1['2016 Vehicles per Household'].cumsum()/df_1["2016 Vehicles per Household"].sum()*100

df_1.reset_index(inplace=True)
df_1.head(10)
```

```python
trace0 = go.Bar(
    x=df_1['states'],
    y=df_1['2016 Vehicles per Household'],
    text=df_1['2016 Vehicles per Household'],
    #textposition = 'auto',
    name='2016 Vehicles per Household'
)

trace1 = go.Line(
    x =df_1['states'],
    y=df_1['percentage'],
    name='Percentage',
    xaxis='x', 
    yaxis='y2'
)

data = [trace0, trace1]
layout = go.Layout(
    title='Top 10 States With The Highest Average Number of Vehicles Per Household In 2016',
    yaxis2=dict(anchor='x', overlaying='y', side='right'),
    legend=dict(x=1,y=1)
)


fig = go.Figure(data=data, layout=layout)

cf.iplot(fig)
```

## Task 2:
**Create a relative frequency distribution histogram from the “2016 percentage of households without vehicles” for the entire country. Comment on the shape of the distribution.**

```python
df_2=uscars[['Jusidiction','2016 Households Without Vehicles']]

df_2_1=df_2.groupby('2016 Households Without Vehicles').count()

df_2_1['frequency']=df_2_1['Jusidiction']
df_2_1.drop('Jusidiction',axis=1,inplace=True)
df_2_1['relative frequency']=(df_2_1['frequency']/(df_2_1["frequency"].sum()))*100
df_2_1.head(10)
```

```python
df_2_1.reset_index(inplace=True)
df_2_1.head(10)

trace2 = go.Histogram(
    x=df_2_1['2016 Households Without Vehicles'],
    y=df_2_1['relative frequency'],
    histnorm='probability',
    #mode='markers'
)
data=[trace2]
layout = go.Layout(
    title='Relative Frequency of 2016 Households Without Vehicles',
    yaxis=dict(title='Relative Frequency'),
    xaxis=dict(title='2016 Households Without Vehicles')
)

fig = go.Figure(data=data,layout=layout)
cf.iplot(fig)
```

## Task 3:
**Create a cumulative frequency line plot of the “2016 percentage of households without vehicles” for the entire country.**

```python
df_2_1['cumulative frequency']=df_2_1['frequency'].cumsum()
df_2_1.set_index('2016 Households Without Vehicles',inplace=True)
df_2_1.head(10)
```

```python
df_2_1['cumulative frequency'].iplot(kind='line',
                                     title='Cumulative Frequency of The “2016 Percentage of Households Without Vehicles”')
```

## Task 7:
**Create a table that shows the average number of vehicles per household by state in 2016 (organize the table by states in alphabetic order). Next, create a same table for 2015, and create a scatter plot of the average number of vehicles per household by state in 2016 versus that of 2015. Use the scatter plot to describe whether there exists a correlation between the two quantities above. Interpret this correlation in words.**

**1.the average number of vehicles per household by state in 2016**

```python
df=uscars[['Jusidiction','2016 Vehicles per Household']]
df['states']=df['Jusidiction'].apply(lambda s:s.split(',')[-1])

ave_2016=df.groupby('states').mean().sort_values(by='states')
ave_2016.head(10)
```

**2.the average number of vehicles per household by state in 2015**

```python
df=uscars[['Jusidiction','2015 Vehicles per Household']]
df['states']=df['Jusidiction'].apply(lambda s:s.split(',')[-1])

ave_2015=df.groupby('states').mean().sort_values(by='states')
ave_2015.head(10)
```

**3.scatter plot (2016 vs 2015)**

```python
ave=pd.merge(ave_2015,ave_2016,on='states',how='inner').sort_values(by='states')
ave.head(10)
```

```python
trace6 = go.Scatter(
    x=ave['2015 Vehicles per Household'],
    y=ave['2016 Vehicles per Household'],
    mode='markers'
)
data=[trace6]
layout = go.Layout(
    title='Correlation Between 2016 and 2015 in Average Number Of Vehicles per Household by State',
    yaxis=dict(title='2016 Vehicles per Household'),
    xaxis=dict(title='2015 Vehicles per Household')
)

fig = go.Figure(data=data,layout=layout)
cf.iplot(fig)
```
